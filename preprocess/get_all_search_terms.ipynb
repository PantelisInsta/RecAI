```python
import os
import pandas
```

```python
# Define save folder
save_folder = os.path.join(os.pardir, 'resources', 'groceries')
print(save_folder)
```

```python
# load csv file of search terms
search_terms = pandas.read_csv('all_substitute_search_terms_aldi.csv')
search_terms.head(3)
```

```python
# Print the number of rows of dataframe
print(search_terms.shape)
```

```python
# Create column cardinality which contains the number of product ids in the list in PRODUCT_IDS
search_terms['cardinality'] = search_terms['PRODUCT_IDS'].apply(lambda x: len(x.split(',')))
search_terms.head(10)
```

```python
# Find all rows with 'milk' in the column TERM
milk_rows = search_terms[search_terms['TERM'] == 'milk']
# print all milk_rows
print(milk_rows)
print(milk_rows.shape)
```

```python
# If there are duplicates in column TERM, remove the ones with the smallest cardinality
terms = search_terms.sort_values(by=['TERM', 'cardinality'], ascending=[False, False])
# Drop the duplicates with smallest cardinality
terms = terms.drop_duplicates(subset='TERM', keep='first')
# Re-order according to index
terms = terms.sort_index()
print(terms.shape)
terms.head(10)
```

```python
# Drop any rows with cardinality 0
terms = terms[terms['cardinality'] > 0]
print(terms.shape)
```

```python
# Save terms column in a csv file in the save folder
terms['TERM'].to_csv(os.path.join(save_folder, 'search_terms.csv'), index=False)
```
